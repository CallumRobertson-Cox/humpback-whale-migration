Initially looking at the data

```python
import pandas as pd
```

```python
data = pd.read_csv("data.csv")
data.head()
```

Sampling data

```python
from imblearn.over_sampling import RandomOverSampler
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
```

```python
train, val, test = np.split(data.sample(frac =1), [int(0.6*len(data)), int(0.8*len(data))])
```

```python
def plot_loss(history):
    plt.plot(history.history["loss"], label = "loss")
    plt.plot(history.history["val_loss"], label = "val_loss")
    plt.xlabel("Epoch")
    plt.ylabel("MSE")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def scale_set(dataframe, oversample = False):
    #actual values
    x = dataframe.drop(columns=["location-lat", 'location-long']).values
    y = dataframe[['location-long', 'location-lat']].values


    #take x and tranform all the values 
    scalar = StandardScaler()
    x = scalar.fit_transform(x)
    
    data = np.hstack((x, y))

    return data, x, y
```

```python
_, X_train, y_train = scale_set(train, oversample = True)
_,X_valid,y_valid = scale_set(val, oversample = False)
_,X_test, y_test = scale_set(test, oversample = False)
```

```python
nn = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])
```

```python
nn.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.1), loss='mean_squared_error')
```

```python
history = nn.fit(
    X_train, y_train,
    epochs = 100,
    validation_data = (X_valid, y_valid)
)
```

```python
plot_loss(history)
```

```python
nn = nn = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation = 'relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(64, activation = 'relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(2, activation = 'linear')
])
```

```python
nn.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.1), loss='mean_squared_error')
```

```python
history = nn.fit(
    X_train, y_train,
    epochs = 100,
    validation_data = (X_valid, y_valid)
)
```
